#**1.** 스마트 스피커
---
### 첫 번째로 음성 분야입니다. 음성 분야에서 제가 소개할 제품은 **스마트 스피커**입니다. 스마트 스피커는 오늘날 주변에서 많이 볼 수 있습니다. 아마존의 에코(Echo), 구글의 구글홈 (Google home)이 있고, 한국 제품으로 기가 지니, WAVE, 카카오 미니가 있습니다. 스마트 스피커는 간단히 말을 알아들을 수 있어서 상호작용이 가능한 스피커라고 할 수 있습니다. 스마트 스피커는 입력 기기이자 출력 기기입니다. 음성으로 마우스나 키보드를 대신하여 장치에 입력을 할 수 있으며, 다양한 출력 서비스를 지원합니다. 이런 서비스는 검색, 노래 재생, 일기예보, 전화, 가전제품의 세부조절 등이 있습니다.
### 스마트 스피커의 핵심 기술은 음성을 듣고, 이해하고, 수행하는 데 있습니다. 음성인식은 사용자의 음성 언어를 컴퓨터가 이해하는 언어로 변화하는 기술입니다. 이렇게 변화된 음성은 컴퓨터가 이해하는 언어로 전환되어 그 의미를 파악합니다. 음성 언어는 다양한 패턴이 있습니다. 예를 들어서 소리를 키워달라는 의미를 “소리 켜줘”, “볼륨 키워줘”, “볼룸 높여” 등 다양한 패턴으로 이야기할 수 있습니다. 이런 다양한 패션을 학습하여 음성 엔진이 이해하여 서비스를 실행합니다. 이때 머신러닝의 하위 분야인 **딥 러닝(deep learning)**이 사용됩니다. 더 많은 데이터를 축척하고, 학습함으로써 더 많은 표현을 이해하고 서비스를 원활하게 실행할 수 있게 됩니다.

#**2.** 번역 서비스
---
### 두 번째는 언어 분야입니다. 언어 분야에서 제가 소개할 제품은 **번역 서비스**입니다. 번역 서비스는 다양한 언어를 자신의 원하는 언어로 변환시켜주는 서비스입니다. 대표적으로 구글 번역(Google Translate), 마이크로소프트의 스카이프 트랜스레이터(Skype Translator), 네이버의 파파고(papago)가 있습니다.
### 파파고의 번역 기술에 관해서 이야기해보겠습니다. 파파고는 인공신경망을 기반으로 하는 **기계번역 기술**을 이용합니다. 인공신경망 기계번역 기술은 대용량의 데이터와 정보, 학습에 필요한 변수 정보를 토대로 기계가 알아서 학습하고, 이후 사람들이 새로운 정보를 추가시켜주었을 경우 기존의 학습에 맞추어 가장 적합한 결과를 출력하는 기술입니다. 파파고는 입력 문장을 하나의 단위로 보고. 전체 맥락을 파악합니다. 그리고 기존의 학습된 데이터 토대로 가장 적합한 문장으로 번역을 해줍니다. 이 기술로 인하여 파파고고는 문장을 쪼개어 번역하던 기존의 방식보다 자연스럽고, 가독성이 좋은 번역 서비스를 제공할 수 있게 되었습니다. 이러한 번역 기술은 앞으로 더 많은 데이터와 다양한 신경망 기반 모델을 가지고 발전될 것입니다. 앞으로 세계의 수많은 다양한 언어를 해석하는 서비스를 제공하게 될 것입니다. 그래서 번역을 통해 문화를 공유되는 사회가 올 것으로 예상됩니다.


#3. 구글렌즈
### 세 번째는 이미지 분석 분야입니다. 이미지 분석 기술에서 살펴볼 기술은 **‘구글 렌즈’**입니다. 구글 렌즈는 스마트폰에서 카메라 기능을 이용하여 사제품, 텍스트 등 물체를 스캔하여 그 대상에 대해서 알려줍니다. 단어를 번역할 수 있고, 동, 식물 이름을 찾고, 제품의 정보를 알려줍니다. 구글 렌즈에서 핵심 기술은 카메라를 통해서 물체를 인식하게 되는 것입니다.
### 이 기술이 가능한 이유는 이는 대규모 데이터 셋과 컴퓨팅 리소스의 출현으로  **CNN(Convolutional Neural Network)** 기술이 발전했기 때문입니다. CNN 기술은 기계학습을 통하여 이미지를 처리하기 위해 개발된 심층신경망입니다. 구글 렌즈와 같이 물체를 인식하고 분류작업을 수행하는 데 주로 사용합니다. CNN 기술은 입력 이미지에서 핵심 정보를 추출합니다. 그리고 여러 층을 거치며 학습을 통해 핵심정보를 인식합니다. 그리고 분류를 진행하게 됩니다. 이러한 학습을 통해서 찍은 사진의 정보를 구글 렌즈는 알려주는 것입니다.


#**4.**자율주행 자동차
---
### 네 번째는 자율주행이 분야입니다. 자율주행이 분야에서 대표적인 것은 자동차입니다. **자율주행 자동차**는 운전자의 개입 없이 스스로 주행이 가증한 자동차를 말합니다. 자율주행이 기술은 국제자동차 기술자협회에서 규정한 6단계 수준으로 구분됩니다. 한국에서는 아직 주행을 보조해주는 기술단계인 Level 2밖에 상용되지 않았습니다. 하지만 테슬라, 아우디, 구글 등 유명 업계에서 반자율주행 단계인 level 3의 단계까지 개발이 완료되었으며, 상용화까지 얼마 남지 않았다고 합니다.
### 자율주행자동차는 복합적인 기술을 요구합니다. 자동차가 스스로 주행하기 위한 절차로 주행 환경 인지, 주행 상황 판단, 차량 제어의 세 가지 단계로 구성됩니다. 이 모든 과정에서 딥러닝이 사용될 수 있습니다. 하지만 저는 주행 상황 판단과 **RNN(Recurrent Neural Network) 기술**에 관해서 이야기하겠습니다. RNN 기술은 심층 신경망 기술 중의 하나로, 내부 노드가 순환 구조로 연결된 형태입니다. 어느 특정 부분이 반복되는 구조로 인하여 순서를 학습하기 효과적인 방법입니다. 데이터의 규칙적 패턴을 학습하여 과거의 정보를 통해 현재의 정보를 파악할 수 있게 해줍니다. 이러한 자율주행이 자동차에서는 실시간으로 장면을 이해하는 기법을 적용할 수 있습니다. 최종적으로 이를 통해 주행 상황 및 주변 차량의 주행 의도를 분석할 수 있습니다.
